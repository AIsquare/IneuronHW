**In this assignment students have to transform iris data into 3 dimensions
and plot a 3d chart with transformed dimensions and colour each data
point with specific class.**


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
#from sklearn import datasets
import seaborn as sns
import pandas as pd
```

```python
data = sns.load_dataset('iris')
```

```python
data.head()
```

```python
data.drop('species',inplace=True,axis=1)
```

```python
data
```

```python
xs = data['sepal_length']
xs.head()
```

```python
ys = data['sepal_width']
```

```python
zs = data['petal_length']
```

```python
z1 = data['petal_width']
```

```python
fig = plt.figure(figsize=(20,10))
ax = Axes3D(fig, elev=-170, azim=120)

ax.scatter(xs, ys, zs, zdir='z', cmap='virdis', edgecolor='b', s=40,linewidth=0.5)
ax.set_xlabel('Sepal_length')
ax.set_ylabel('Sepal_width')
ax.set_zlabel('Petal_lenght')
plt.show()
```

```python
fig = plt.figure(figsize=(20,10))
ax = Axes3D(fig, elev=-170, azim=120)

ax.scatter(xs, ys, z1, zdir='z', cmap='virdis', edgecolor='b', s=40,linewidth=0.5)
ax.set_xlabel('Sepal_length')
ax.set_ylabel('Sepal_width')
ax.set_zlabel('Petal_width')
plt.show()
```
